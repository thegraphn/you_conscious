```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/home/graphn/repositories/you_conscious/dl_xp/data/colour/colors.csv",sep=";",low_memory=False)
```

```python
texts = df["text"].tolist()
labels = df["label"].tolist()
```

```python
from collections import Counter
print(Counter(labels))

set_labels = list(set(labels))
```

```python
set_labels
```

```python
import sklearn
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(texts,labels, test_size=0.15, random_state=42, shuffle=True, stratify=labels)
```

```python
d_train = pd.DataFrame({"text":X_train,"label":y_train})
d_test = pd.DataFrame({"text":X_test,"label":y_test})
```

```python
d_train.to_csv("/home/graphn/repositories/you_conscious/dl_xp/data/colour/train_color.tsv",sep="\t")
d_test.to_csv("/home/graphn/repositories/you_conscious/dl_xp/data/colour/test_color.tsv",sep="\t")
```
